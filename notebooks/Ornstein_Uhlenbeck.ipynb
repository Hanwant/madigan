```python
import random
import math
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

figsize(9, 6)
```

```python
def OU(mean=10., theta=0.5, phi=0.5, noise_var=1.):
    x = mean
    dt = 1.
    while True:
        dx = theta*(mean-x)*dt + phi*random.gauss(0., noise_var)
        x += dx
        yield x
```

```python
%matplotlib notebook

ou_process = iter(OU(mean=10., theta=.1, phi=4., noise_var=.1))

series = []
for i in range(640):
    series.append(next(ou_process))
    
plt.plot(series)
```
